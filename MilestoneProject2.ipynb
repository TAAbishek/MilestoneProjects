```python
#Global Variables

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10,'Ace1':1,'Ace11':11}
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
    
    def dealer_value(self):
        if self.rank=="Ace":
            return 0
        else:
            return values[self.rank]
    
    def __str__(self):
        return(f"{self.rank} of {self.suit}")
    
c1=Card("Hearts","King")
print(c1.dealer_value())
print(c1)

```

```python
import random
from random import shuffle

class Deck():
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                card=Card(suit,rank)
                self.all_cards.append(card)
                
    def shuffler(self):
        shuffle(self.all_cards)
        
    def remove(self):
        return self.all_cards.pop(0)
        
    def __len__(self):
        return len(self.all_cards)
        
    
d1=Deck()

print(len(d1))
```

```python
class Dealer():
    def __init__(self):
        self.all_cards=[]
        
    def add_cards(self,card):
        self.all_cards.append(card)
    def __len__(self):
        return len(self.all_cards)
    
    def value(self):
        val=0
        card_no=0
       
        for i in self.all_cards[:2]:
            if card_no==0:
                card1=i
            elif card_no==1:
                card2=i
            card_no+=1
        if card1.rank=="Ace" and card2.rank=="Ace":
            val=12
        elif card1.rank=="Ace":
            val=val+11+card2.dealer_value()
        elif card2.rank=="Ace":
            val=val+11+card1.dealer_value()
        else:
            val+=card1.dealer_value()+card2.dealer_value()
        
        for i in self.all_cards[2:]:
            if i.rank=="Ace":
                if val<=10:
                    val+=11
                else:
                    val+=1
            else:
                val+=i.dealer_value()
                
        return val
    
    
```

```python
class Player(Dealer):
    def __init__(self,name):
        self.balance=1000
        self.name=name
        self.all_cards=[]
    
    def add_cards(self,card):
        self.all_cards.append(card)
    
```

```python
def countinue():
    char=input("Countinue? (Y/N)")
    i=False
    while i!=True:
        if char in ["Y","y"]:
            return True
        elif char in ["N","n"]:
            print("Thank you for playing the Game!!")
            return False
        else:
            char=input("incorrect input enter again")
```
